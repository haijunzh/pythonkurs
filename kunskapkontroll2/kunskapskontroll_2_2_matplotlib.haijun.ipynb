### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?






1. Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style"). <br>
Advantages:
* More explicit and structured
* Better for complex plots and larger codebases

2. Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.<br>
* Easy and quick

* Common in simple scripts and interactive use (e.g. notebooks)



The approach 1 is recommended, especially for complex plotting, while approach 2 is fine for quick, simple plots

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.



- Figure <br>
The whole figure. The Figure keeps track of all the child elements (titles, child axes, etc.)<br>
- Axes<br>
An Axes is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects (be aware of the difference between Axes and Axis) that provide ticks and tick labels to provide scales for the data in the Axes. Each Axes also has a title (set via set_title()), an x-label (set via set_xlabel()), and a y-label set via set_ylabel()).<br>
- Axis<br>
These objects set the scale and limits and generate ticks (the marks on the Axis) and ticklabels (strings labeling the ticks).<br>
- Artist<br>
Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects). 


### When plotting in Matplotlib, what is the expected input data type?

Plotting functions expect numpy.array or numpy.ma.masked_array as input, or objects that can be passed to numpy.asarray. Classes that are similar to arrays ('array-like') such as pandas data objects and numpy.matrix may not work as intended. Common convention is to convert these to numpy.array objects prior to plotting. 

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Approach 1: Explicit object-oriented approach

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
ax.plot(x, x**2)
ax.set_title('y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
# Approach 2: Implicit approach using pyplot state machine
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 100)
plt.figure(figsize=(8, 4), layout='constrained')
plt.plot(x, x**2)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), layout='constrained')
ax1.scatter(random_data_x, random_data_y, alpha=0.5)
ax1.set_title('Scatter Plot of Random Data')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax2.bar(names, values, color='orange', alpha=0.7)
ax2.set_title('Fruit Count')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Count')
plt.show()
```
